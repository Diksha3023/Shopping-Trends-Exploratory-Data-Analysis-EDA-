<div style='background-color: #fffec8;padding:2px; color: #53682E;border : 3px solid #000000; font-size: 15px;text-align: center'>
<p>
- <b>Customer ID</b>- Unique Identifier for each customer.<br>
- <b>Age</b>- Age of the customer.<br>
- <b>Gender</b>- Gender of the customer(Male/Female).<br>
- <b>Item Purchased</b>- The item purchased by the customer.<br>
- <b>Category</b>- Category of the item purchased.<br>
- <b>Purchase Amount (USD</b>- The amount of the purchase in USD.<br>
- <b>Location</b>- Location where the purchase was made.<br>
- <b>Size</b>- Size of the purchased item.<br>
- <b>Color</b>- Color of the purchased item.<br>
- <b>Season</b>- Season during which the purchase was made.<br>
- <b>Review Rating</b>- Rating given by the customer for the purchased item.<br>
- <b>Subscription Status</b>- Indicates if the customer has a subscription(Yes/No).<br>
- <b>Shipping Type</b>- Type of shipping chosen by the customer.<br>
- <b>Discount Applied</b>- Indicates if a discount was applied to the purchase(Yes/NO).<br>
- <b>Promo Code Used</b>- Indicates ig a promo code was used for the purchase(Yes/No).<br>
- <b>Previous Purchase</b>- Number of previous purchases made by the customer.<br>
- <b>Payment Method</b>- Customer's most preferred payment method.<br>
- **Frequency of Purchases</b>- Frequency at which the customer makes purchases(e.g.- Weekly, Fortnightly,Monthly).</p>
</div>

<p>&nbsp;
Find out the Gender distribution in Top 10 locations with most Purchases? Also find the most purchased items &nbsp;by male and female in those locations.
</p>
</div>

```python
location1=pd.DataFrame(df.groupby(["Location","Gender"]).size().unstack())
location1=location1.assign(sum= df.groupby("Location").size()).sort_values(by='sum',ascending=True).tail(10).reset_index()
dd=df[df["Location"].isin(df.groupby("Location").size().sort_values(ascending=False).head(10).index)]
a=dd.groupby(["Location","Gender","Item Purchased"]).size()
b=pd.DataFrame(a)
b=b.assign(New=df.groupby(["Location","Gender","Item Purchased"]).size())
c=b.groupby(["Location","Gender"])["New"].idxmax()
location2=b.loc[c].reset_index()
print(location2)
location=pd.merge(location1,location2,on="Location")
x1=location[location["Gender"]=="Female"]["Item Purchased"]
x2=location[location["Gender"]=="Male"]["Item Purchased"]
ax=location1.plot(x="Location", y=["Male",'Female'], kind="barh",stacked=True, figsize=(10,10),color=["#8fb9aa","#f2d096"])
for i,bar in enumerate(ax.patches):
    ax.text(bar.get_x()+bar.get_width()-3,bar.get_y()+bar.get_height()-0.3, "("+str(bar.get_width())+")", ha='center')
    if i%2==0:
        ax.text(location["Female"][i]/2+location["Male"][i],i/2, x1[i], ha='center')
    else:
        ax.text(location["Male"][i]/2,i//2,x2[i],ha="center")
```
